```python
'数据读取与分割 制作dataloader 注意读取的时npy还是csv'
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader


load_data = np.load(trainset_config['train_data_path'])

# np.random.seed(10)
np.random.shuffle(load_data)
train_data = load_data
# train_data = load_data[ 0: int( load_data.shape[0]*0.8) ]
# test_data = load_data[ int( load_data.shape[0]*0.8): ]
# print(train_data.shape)


Number, Length, Attribute  = train_data.shape
print(Number, Length, Attribute)


class MyDataset(Dataset):
    def __init__(self, X_data):
        self.X_data = X_data

    def __len__(self):
        return len(self.X_data)

    def __getitem__(self, idx):
        x = torch.tensor(self.X_data[idx], dtype=torch.float32)  # 转换为 Tensor
        return x   
                    
dataset = MyDataset(train_data)    
train_dataloader = DataLoader(dataset, batch_size=200, shuffle=True)
print(next(iter(train_dataloader)).shape)
# print(next(iter(train_dataloader))[0])

# for batch in train_dataloader:
#     print(batch.shape)
```

```python
def resfusion_x0_to_xt(x_0: torch.Tensor, alpha_hat_t: torch.Tensor, residual_term: torch.Tensor, noise: torch.Tensor) \
        -> torch.Tensor:
    """
    Compute x_t from x_0 with formula 20 from Resfusion
    :param x_0: the image without noise
    :param alpha_hat_t: the cumulated variance schedule at time t
    :param residual_term: the residual term
    :param noise: pure noise from N(0, 1)
    :return: the noised image x_t at step t
    """
    return (torch.sqrt(alpha_hat_t) * x_0 + (1 - torch.sqrt(alpha_hat_t)) * residual_term
            + torch.sqrt(1 - alpha_hat_t) * noise)


def ddpm_x0_to_xt(x_0: torch.Tensor, alpha_hat_t: torch.Tensor, noise: torch.Tensor) -> torch.Tensor:
    """
    Compute x_t from x_0 using a closed form using theorem from original DDPM paper (Ho et al.)
    :param x_0: the image without noise
    :param alpha_hat_t: the cumulated variance schedule at time t
    :param noise: pure noise from N(0, 1)
    :return: the noised image x_t at step t
    """
    return torch.sqrt(alpha_hat_t) * x_0 + torch.sqrt(1 - alpha_hat_t) * noise
```

```python
epoch = 500   # 基本上500足够 好像energy 的res和skip channels在512时需要1000
mr = 0.3
learning_rate = 0.0002
missing_length = int( mr * Length )


loss_list=[]
net = SSSDS4Imputer(**model_config).cuda()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)

print_size(net)
```

```python
'这里训练时选择的是全局加噪 可以考虑只在缺失处加噪 参考sssd 这样的话在训练和采样时 要保证观测处不变 只在缺失处加减噪 同时只计算缺失处的损失'


for e in tqdm(range(epoch)):
      
    for batch in train_dataloader:   # [128, 30, 6]
        
        X_0 = batch.float().cuda()  
        B, L, D = X_0.shape


        '4种缺失机制  掩码可全相同也可不同'
        '掩码都相同'
             
# #         mask = get_mask_rm( X_0[0], missing_length)      # 原数据乘以缺失掩码
# #         mask = get_mask_mnr( X_0[0], missing_length) 
# #         mask = get_mask_bm( X_0[0], missing_length) 
#         mask = get_mask_mcar( X_0[0], mr)  # mcar缺失法的输入参数和另外单个不一样 直接输入缺失率
    
#         mask = mask.repeat(X_0.size()[0], 1, 1).float().cuda()
        
        
        '产生不同掩码'
        mask = torch.empty([0, Length, Attribute])
        
        for i in range(B):
#             one_mask = get_mask_mcar( X_0[0], mr)  
            one_mask = get_mask_mnr( X_0[0], missing_length)   # mnr由于其算法问题 缺失时起始点的不同 导致每一列的缺失数量不一定相同 基本会小于设定的缺失率
#             one_mask = get_mask_rm( noise[0], missing_length)      
#             one_mask = get_mask_bm( X_0[0], missing_length)  
            
            one_mask = one_mask.unsqueeze(0)
            mask = torch.row_stack([mask, one_mask])

        mask = mask.float().cuda()
        
         
        X_0_hat = X_0 * mask

        residual_term = X_0_hat - X_0 
        
        t: torch.Tensor = torch.randint(0, T_acc, (X_0.shape[0],) ).cuda()     #  
        alpha_hat = Alpha_bar[t].reshape(-1, 1, 1).cuda()                      # 
        noise = torch.randn_like(X_0)
        
        
        
        x_t = resfusion_x0_to_xt(X_0, alpha_hat, residual_term, noise) 
        
        '局部加噪 使用局部加噪时 采样时也要保证观测处一致 只在缺失区域处理'
        x_t = x_t*(1-mask) + X_0*mask
        
        
        
        alpha = Alpha[t].reshape(-1, 1, 1).cuda() 
        beta = Beta[t].reshape(-1, 1, 1).cuda() 
        
        resnoise = noise + (1 - torch.sqrt(alpha)) * torch.sqrt(1 - alpha_hat) / beta * residual_term
        
    
        '两种训练模式 有条件和无条件 有条件是融入观测值和掩码 二者会拼接'
        '第二个参数是完整数据 输入模型后会和掩码写成 变成观测值 是用观测值作为条件 而不是完整数据作为条件 这种写法有些迷惑性 直接写成X_0*mask更更好'
        '这里和sssd里的全局加噪方式一致 可以考虑只对缺失位置加噪'
        pred_resnoise = net( (x_t.permute(0, 2, 1), X_0_hat.permute(0, 2, 1), mask.permute(0, 2, 1), t.view(B, 1),) )  #  [B, 6, 30]
#         pred_resnoise = net( (x_t.permute(0, 2, 1), None, None, t.view(B, 1),) ) 
        pred_resnoise = pred_resnoise.permute(0, 2, 1)
        
        optimizer.zero_grad()
        
        
        
#         全局mse 不考虑观测部分
#         loss = F.mse_loss(input=pred_resnoise, target=resnoise ) 
        
    
        '只计算观测部分的损失 相比全局mse loss在最开始位置的收敛会比较平滑 最终效果也比较好 '
        '实现方法1  直接除以mr在mcar情况下和方法2一致 但是在mnr情况下不够合理 因为mnr下真实缺失率往往小于mr 所以要除以真实缺失率 即直接用掩码计算 见第二行'
#         loss = F.mse_loss(input=pred_resnoise*(1-mask), target=resnoise*(1-mask) ) / mr
#         loss = F.mse_loss(input=pred_resnoise*(1-mask), target=resnoise*(1-mask) )/ ( (1-mask).sum()/ ( (1-mask).sum() +mask.sum())  )
   
        
        '实现方法2'
        '这个方法更精确 特别是对于mnr情况 时间的缺失数量往往小于设定的mr 所以方法1里的直接除以mr不够合理 而该方法较合理'
        loss = F.mse_loss(input=pred_resnoise*(1-mask), target=resnoise*(1-mask),  reduction='none' )
        loss = loss.sum()/(1-mask).sum()
        
        # 查看实际缺失率
#         print( (1-mask).sum()/( (1-mask).sum()+mask.sum() ) )    
#         print( (1-mask).sum()/( B* L*D )  )

    
#         RMSE loss
#         loss = torch.sqrt( F.mse_loss(input=pred_resnoise(1-mask), target=resnoise(1-mask) )/ mr )  # 损失可以除以batchsize
#         smooth_l1 loss
#         loss = F.smooth_l1_loss(input=pred_resnoise, target=resnoise)
    
    
        loss.backward()
        optimizer.step()
        loss_list.append(loss.item())
    epoch += 1
```

```python
class Reflow_Dataset(Dataset):
    def __init__(self, X_data,Y_data, A_data, B_data ):
        self.X_data = X_data
        self.Y_data = Y_data
        self.A_data = A_data
        self.B_data = B_data

    def __len__(self):
        return len(self.X_data)
    
    def __getitem__(self, idx):
        x = torch.tensor(self.X_data[idx], dtype=torch.float32)  # 转换为 Tensor
        y = torch.tensor(self.Y_data[idx], dtype=torch.float32)
        a = torch.tensor(self.A_data[idx], dtype=torch.float32)
        b = torch.tensor(self.B_data[idx])
        return x,y,a,b
    
                 
'reflow数据集分别包含ddim采样的x0 加噪的观测值x_Tacc 观测值X_0_hat 掩码'
reflow_dataset = Reflow_Dataset(DDIM_res.permute(0, 2, 1), X_Tacc*(1-Mask)+X_0_hat*Mask, X_0_hat, Mask )   
reflow_dataloader = DataLoader(reflow_dataset, batch_size=200, shuffle=False)
print(len( next(iter(reflow_dataloader) )) )
print(next(iter(reflow_dataloader))[2].shape)
```

```python
for e in tqdm(range(reflow_epoch)):
      
    for batch in reflow_dataloader:   # [128, 30, 6]
        
        x_1 = batch[0].float().cuda()      # resnoise生成的原数据 
        x_0 = batch[1].float().cuda()      # 缺失数据的加噪值 采样起点
        x_0_hat = batch[2].float().cuda() # 观测数据  作为条件
        mask = batch[3].float().cuda()
        
        B, L, D = x_0.shape
        
        
        '开始整流加噪 应该是对原数据加噪 x1是原图 x0是终点'
        t = torch.randint(min_step, max_step + 1, (x_0.shape[0],) ).cuda()          
        t_view = t.view(B, 1, 1)
        x_t = ( t_view/T_acc ) * x_1 + ( 1-t_view/T_acc ) * x_0 # 整流加噪时t缩放到0-1
        x_t = x_t * (1-mask) + x_1 * mask
        
        '预测时t的2个使用方式  可以用离散的 也可以转换为0-1再用  第一个方法收敛更快' 
        pred_v = student_unet( (x_t.permute(0, 2, 1), x_0_hat.permute(0, 2, 1), mask.permute(0, 2, 1), t.view(B, 1),) )   
#         pred_v = student_unet( (x_t, None, None, t.view(B, 1),) )                   
#         pred_v = student_unet( (x_t, 1, 1, t.view(B, 1)/T_acc,) )  
             
        
        '注意loss里x0和x1的位置  用x0-x1时loss很小 可达0.002以下 可收敛 到后面会缓慢减小'
        '用x1-x0时loss较大 为0.7多一点 也可以收敛'
#         loss = F.mse_loss(pred_v, (X_1 - X_0.permute(0, 2, 1) ))
#         loss = F.mse_loss(pred_v, ( X_0.permute(0, 2, 1)- X_1 ), reduction="mean")

#         loss = F.mse_loss(pred_v, ( x_0- x_1 ).permute(0, 2, 1))
        
        
        loss = F.mse_loss(input=pred_v.permute(0, 2, 1)*(1-mask), target=( x_0-x_1 )*(1-mask), reduction='none' )
        loss = loss.sum()/(1-mask).sum()
        
        
        optimizer.zero_grad()
        loss.backward()
#         torch.nn.utils.clip_grad_norm_(student_unet.parameters(), 1.0)
        optimizer.step()
        print(loss.item())
        
        loss_list.append(loss.item())
           
    e += 1
        
```

```python
'制作distillation数据集'

class Distill_Dataset(Dataset):
    def __init__(self, X_data,Y_data, A_data, B_data ):
        self.X_data = X_data
        self.Y_data = Y_data
        self.A_data = A_data
        self.B_data = B_data

    def __len__(self):
        return len(self.X_data)
    
    def __getitem__(self, idx):
        x = torch.tensor(self.X_data[idx], dtype=torch.float32)  # 转换为 Tensor
        y = torch.tensor(self.Y_data[idx], dtype=torch.float32)
        a = torch.tensor(self.A_data[idx], dtype=torch.float32)
        b = torch.tensor(self.B_data[idx])
        return x,y,a,b
    
                 
'reflow数据集分别包含ddim采样的x0 加噪的观测值x_Tacc 观测值X_0_hat 掩码'
distill_dataset = Distill_Dataset(reflow_res.permute(0, 2, 1), X_Tacc*(1-Mask)+X_0_hat*Mask, X_0_hat, Mask )   
distill_dataloader = DataLoader(distill_dataset, batch_size=200, shuffle=False)
print(len( next(iter(distill_dataloader) )) )
print(next(iter(distill_dataloader))[2].shape)
```

```python
from scipy.integrate import quad

# 计算两个值之间的速度  用两个值之差 除以时间间隔 得到速度  
def compute_velocity(z_t, z_t_minus_1, delta_t):
    v_t = (z_t_minus_1 - z_t) / delta_t
    return v_t
                         # (248, 1, 4, 32, 32)  上面的函数     false
def straightness_measurement(Z_t_list, compute_velocity, use_velocity_fn):
    """
    Measure the straightness of a trajectory given initial and final positions (Z_0, Z_1)
    and a velocity function v(Z_t, t).
    """

#     Z_0 = Z_t_list[0]   # 估计的起点和终点
#     Z_1 = Z_t_list[-1]
    
    
    Z_0 = Z_t_list[-1]   # 修改方向 这个要改的话  上面的compute_velocity里的两个相减的值就不用换位置
    Z_1 = Z_t_list[0]
    

    def integrand(t):
        # t is [0, 1], we must scale it to the length of Z_t_list
        i = t * (len(Z_t_list) - 1)
        delta_t = 1 / len(Z_t_list) # 时间间隔 for now we assume linearly spaced time steps

        if use_velocity_fn:            # 当前为false 不用额外计算向量
            if t == 0:
                v_t = 0
            else:                  # 计算相邻两点直接的速度
                v_t = compute_velocity(Z_t_list[int(i)], Z_t_list[int(i-1)], delta_t)  
        else:
            v_t = Z_t_list[int(i)]     # 直接取向量值
            
        deviation = (Z_1 - Z_0) - v_t
#         deviation = np.array(deviation.cpu())  #要转换成np格式才好进行下面的取均值计算
        
        return np.mean(deviation ** 2)

    # Integrate from t=0 to t=1
    S_Z, _ = np.log(quad(integrand, 0, 1))
    return S_Z


straightness_value_ddpm = straightness_measurement( trajectory_ddpm[0:-1], compute_velocity, use_velocity_fn=True)
straightness_value_ddim = straightness_measurement( trajectory_ddim[0:-1], compute_velocity, use_velocity_fn=True)
straightness_value_reflow = straightness_measurement( trajectory_reflow[0:-1], compute_velocity, use_velocity_fn=True)
print(straightness_value_ddpm, straightness_value_ddim, straightness_value_reflow )
```
