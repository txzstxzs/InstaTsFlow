```python
for key in diffusion_hyperparams:
    if key != "T":
        diffusion_hyperparams[key] = diffusion_hyperparams[key].cuda()
        
dh = diffusion_hyperparams
T, Alpha, Alpha_bar, Sigma = dh["T"], dh["Alpha"], dh["Alpha_bar"], dh["Sigma"]
Beta = 1 - Alpha
print(T)

def get_acc_point(alphas_hat):
        """
            Calculate Acceleration Point, according to formula 27
        """
        abs_dist = torch.abs(torch.sqrt(alphas_hat) - 0.5)
        return abs_dist.argmin() + 1
    
T_acc = get_acc_point(Alpha_bar).item()
print( T_acc )

abs_dist = torch.abs(torch.sqrt(Alpha_bar[165]) - 0.5)
```

```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

load_data = np.load(trainset_config['train_data_path'])
np.random.shuffle(load_data)
train_data = load_data

Number, Length, Attribute  = train_data.shape
print(Number, Length, Attribute)

class MyDataset(Dataset):
    def __init__(self, X_data):
        self.X_data = X_data

    def __len__(self):
        return len(self.X_data)

    def __getitem__(self, idx):
        x = torch.tensor(self.X_data[idx], dtype=torch.float32)  
        return x   
                    
dataset = MyDataset(train_data)    
train_dataloader = DataLoader(dataset, batch_size=200, shuffle=True)
print(next(iter(train_dataloader)).shape)
```

```python
def resfusion_x0_to_xt(x_0: torch.Tensor, alpha_hat_t: torch.Tensor, residual_term: torch.Tensor, noise: torch.Tensor) \
        -> torch.Tensor:
    
    return (torch.sqrt(alpha_hat_t) * x_0 + (1 - torch.sqrt(alpha_hat_t)) * residual_term
            + torch.sqrt(1 - alpha_hat_t) * noise)

def ddpm_x0_to_xt(x_0: torch.Tensor, alpha_hat_t: torch.Tensor, noise: torch.Tensor) -> torch.Tensor:
    return torch.sqrt(alpha_hat_t) * x_0 + torch.sqrt(1 - alpha_hat_t) * noise


def std_normal(size):
    return torch.normal(0, 1, size=size).cuda()
```

```python
epoch = 1000   
mr = 0.3
learning_rate = 0.0002
missing_length = int( mr * Length )

loss_list=[]

net = SSSDS4Imputer(**model_config).cuda()
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)
```

```python
for e in tqdm(range(epoch)):
      
    for batch in train_dataloader:   
        
        X_0 = batch.float().cuda()  
        B, L, D = X_0.shape

        mask = torch.empty([0, Length, Attribute])
        
        for i in range(B):
            one_mask = get_mask_mcar( X_0[0], mr)  
#             one_mask = get_mask_mnr( X_0[0], missing_length)  
            one_mask = one_mask.unsqueeze(0)
            mask = torch.row_stack([mask, one_mask])

        mask = mask.float().cuda()
        
         
        X_0_hat = X_0 * mask
        residual_term = X_0_hat - X_0 
        t: torch.Tensor = torch.randint(0, T_acc, (X_0.shape[0],) ).cuda()     
        alpha_hat = Alpha_bar[t].reshape(-1, 1, 1).cuda()                     
        noise = torch.randn_like(X_0)
        x_t = resfusion_x0_to_xt(X_0, alpha_hat, residual_term, noise) 
        x_t = x_t*(1-mask) + X_0*mask
             
        alpha = Alpha[t].reshape(-1, 1, 1).cuda() 
        beta = Beta[t].reshape(-1, 1, 1).cuda() 
        
        resnoise = noise + (1 - torch.sqrt(alpha)) * torch.sqrt(1 - alpha_hat) / beta * residual_term
        

        pred_resnoise = net( (x_t.permute(0, 2, 1), X_0.permute(0, 2, 1), mask.permute(0, 2, 1), t.view(B, 1),) ) 
        pred_resnoise = pred_resnoise.permute(0, 2, 1)
        
        optimizer.zero_grad()
        
    
        loss = F.mse_loss(input=pred_resnoise*(1-mask), target=resnoise*(1-mask),  reduction='none' )
        loss = loss.sum()/(1-mask).sum()
        
    
        loss.backward()
        optimizer.step()
        loss_list.append(loss.item())
    epoch += 1
```

```python
Mask = torch.empty([0, Length, Attribute])    
print( Mask.shape)
print(mr, missing_length)

for i in range(Number):
    one_mask = get_mask_mcar( noise[0], mr)  
#     one_mask = get_mask_mnr( train_data[0], missing_length) 
    one_mask = one_mask.unsqueeze(0)   
    Mask = torch.row_stack([Mask, one_mask])

Mask = Mask.float().cuda()
print(Mask.shape)
    

X_0_hat = train_data.reshape(-1, Length, Attribute) * Mask   
alpha_hat = Alpha_bar[T_acc - 1].reshape(-1, 1, 1).cuda()
X_Tacc = ddpm_x0_to_xt(X_0_hat, alpha_hat, noise)     
print(X_Tacc.shape)
```

# Resnoise sampling

```python
'DDPM sampler'
X = X_Tacc.permute(0, 2, 1)
X = X*(1-Mask.permute(0, 2, 1)) + X_0_hat.permute(0, 2, 1)
print(X.shape)

trajectory_ddpm = []

start_time = time.time()
with torch.no_grad():      
        for t in tqdm( range( T_acc - 1, -1, -1) ):   
                
            trajectory_ddpm.append( X.permute(0, 2, 1).cpu().numpy() )
            diffusion_steps = (t * torch.ones((size[0], 1))).cuda()    
            epsilon_theta = net((X, X_0_hat.permute(0, 2, 1), Mask.permute(0, 2, 1), diffusion_steps,)) 
            
            alpha = Alpha[t].reshape(-1, 1, 1).cuda() 
            beta = Beta[t].reshape(-1, 1, 1).cuda() 
            alpha_hat = Alpha_bar[t].reshape(-1, 1, 1).cuda()      
            sigma = Sigma[t].reshape(-1, 1, 1).cuda()      
            
            X = ( X - ( (1 - alpha) / ( torch.sqrt(1 - alpha_hat) ) * epsilon_theta) )  / torch.sqrt(alpha)
            
            if t > 0:      
                X = X + sigma * std_normal( (Number, Attribute, Length ))
                                
            X = X*(1-Mask.permute(0, 2, 1)) + X_0_hat.permute(0, 2, 1)
    
DDPM_res = X
trajectory_ddpm.append( X.permute(0, 2, 1).cpu().numpy() )
trajectory_ddpm.append( Mask.cpu().numpy() )

end_time = time.time()
execution_time = end_time - start_time   
print(execution_time) 



'DDIM sampler'
X = X_Tacc.permute(0, 2, 1)
X = X*(1-Mask.permute(0, 2, 1)) + X_0_hat.permute(0, 2, 1)

trajectory_ddim = []

start_time = time.time()
with torch.no_grad():      
        for t in tqdm( range(T_acc - 1 , 0, -1) ):    
                
            trajectory_ddim.append( X.permute(0, 2, 1).cpu().numpy() )
            diffusion_steps = (t * torch.ones((size[0], 1))).cuda()
            epsilon_theta = net((X, X_0_hat.permute(0, 2, 1), Mask.permute(0, 2, 1), diffusion_steps,)) 
            
            alpha_hat = Alpha_bar[t].reshape(-1, 1, 1).cuda()    
            alpha_hat_t_minus_1 = Alpha_bar[t-1].reshape(-1, 1, 1).cuda() 

            if t==1:        
                X = (X - torch.sqrt(1-alpha_hat) * epsilon_theta)  / torch.sqrt(alpha_hat)  
                X = X*(1-Mask.permute(0, 2, 1))+X_0_hat.permute(0, 2, 1)
                continue
                
            
            X = ( torch.sqrt( alpha_hat_t_minus_1 ) * (X - torch.sqrt(1-alpha_hat) * epsilon_theta) ) / torch.sqrt(alpha_hat) + \
                                                                                              torch.sqrt(1-alpha_hat_t_minus_1)* epsilon_theta

            
            X = X*(1-Mask.permute(0, 2, 1)) + X_0_hat.permute(0, 2, 1)

DDIM_res = X
trajectory_ddim.append( X.permute(0, 2, 1).cpu().numpy() )
trajectory_ddim.append( Mask.cpu().numpy() )

end_time = time.time()
execution_time = end_time - start_time  
print(execution_time) 
```

```python
class Reflow_Dataset(Dataset):
    def __init__(self, X_data,Y_data, A_data, B_data ):
        self.X_data = X_data
        self.Y_data = Y_data
        self.A_data = A_data
        self.B_data = B_data

    def __len__(self):
        return len(self.X_data)
    
    def __getitem__(self, idx):
        x = torch.tensor(self.X_data[idx], dtype=torch.float32)  
        y = torch.tensor(self.Y_data[idx], dtype=torch.float32)
        a = torch.tensor(self.A_data[idx], dtype=torch.float32)
        b = torch.tensor(self.B_data[idx])
        return x,y,a,b
    
                 
reflow_dataset = Reflow_Dataset(DDIM_res.permute(0, 2, 1), X_Tacc*(1-Mask)+X_0_hat*Mask, X_0_hat, Mask )  

# reflow_dataset = Reflow_Dataset(DDPM_res.permute(0, 2, 1), X_Tacc*(1-Mask)+X_0_hat*Mask, X_0_hat, Mask )  

reflow_dataloader = DataLoader(reflow_dataset, batch_size=200, shuffle=True)
print(len( next(iter(reflow_dataloader) )) )
print(next(iter(reflow_dataloader))[2].shape)
```

```python
for e in tqdm(range(reflow_epoch)):
      
    for batch in reflow_dataloader:   
        
        x_1 = batch[0].float().cuda()     
        x_0 = batch[1].float().cuda()      
        x_0_hat = batch[2].float().cuda() 
        mask = batch[3].float().cuda()
        
        B, L, D = x_0.shape
        
        t = torch.randint(min_step, max_step + 1, (x_0.shape[0],) ).cuda()          
        t_view = t.view(B, 1, 1)
        x_t = ( t_view/T_acc ) * x_1 + ( 1-t_view/T_acc ) * x_0 
        x_t = x_t * (1-mask) + x_1 * mask
        
        pred_v = student_unet( (x_t.permute(0, 2, 1), x_0_hat.permute(0, 2, 1), mask.permute(0, 2, 1), t.view(B, 1),) )   
          
        loss = F.mse_loss(input=pred_v.permute(0, 2, 1)*(1-mask), target=( x_0-x_1 )*(1-mask), reduction='none' )
        loss = loss.sum()/(1-mask).sum()
            
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        loss_list.append(loss.item())

    e += 1
        
```

```python
from scipy.integrate import quad

def compute_velocity(z_t, z_t_minus_1, delta_t):
    v_t = (z_t_minus_1 - z_t) / delta_t
    return v_t
                         
def straightness_measurement(Z_t_list, compute_velocity, use_velocity_fn):
    """
    Measure the straightness of a trajectory given initial and final positions (Z_0, Z_1)
    and a velocity function v(Z_t, t).
    """
    Z_0 = Z_t_list[-1]   
    Z_1 = Z_t_list[0]
    
    def integrand(t):
        
        i = t * (len(Z_t_list) - 1)
        delta_t = 1 / len(Z_t_list) 

        if use_velocity_fn:           
            if t == 0:
                v_t = 0
            else:                  
                v_t = compute_velocity(Z_t_list[int(i)], Z_t_list[int(i-1)], delta_t)  
        else:
            v_t = Z_t_list[int(i)]     
            
        deviation = (Z_1 - Z_0) - v_t

        return np.mean(deviation ** 2)

    S_Z, _ = np.log(quad(integrand, 0, 1))
    return S_Z


straightness_value_ddpm = straightness_measurement( trajectory_ddpm[0:-1], compute_velocity, use_velocity_fn=True)
straightness_value_ddim = straightness_measurement( trajectory_ddim[0:-1], compute_velocity, use_velocity_fn=True)
straightness_value_reflow = straightness_measurement( trajectory_reflow[0:-1], compute_velocity, use_velocity_fn=True)
print(straightness_value_ddpm, straightness_value_ddim, straightness_value_reflow )
```
